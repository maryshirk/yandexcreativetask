## 1. Работа с пропущенными значениями.

Ознакомимся с датасетом. Для этого обратимся к методу head(n), который возращает первые n строк датасета. Изучим столбцы: org_id - идентификатор заведения, city - город, в котором находится заведение (msk или spb), average_bill - средний чек в заведении, rating - рейтинг заведения, rubric - тип заведения. Остальные столбцы - человекочитаемые бинарные фичи заведений (coffee_to_go, food_delivery, breakfast, summer_terrace, wi_fi, business_lunch, payment_by_credit_card, car_park, gift_certificate, karaoke, view_on_ostankino_tower, free_delivery, music, pets, air_conditioning).

```python
import pandas as pd
import numpy as np
ds = pd.read_csv('datasets_eat_places.csv')
ds.head(10)
```

```python
ds.info()
```

Метод info( ) - показывает информацию о наборе данных, индекс, столбцы и тип данных, ненулевые значения и использование памяти. В результате мы видим, что все колонки, кроме колонок 'average_bill' (17820), 'rating' (36438), содержат по 43195 строк. Следовательно имеем 43195 - 17820 + 43195 - 36438 = **32132** пропущенных значений.

Очистим датасет от всех заведений, у которых средний касовый чек неизвестен или превышает 2500.

```python
delds = ds[(ds.average_bill > 2500) | (np.isnan(ds.average_bill))]
ds = ds.drop(delds.index, axis=0)
```

```python
ds
```

Проверим количество пропусков.

```python
ds.isnull().sum()
```

Заполним оставшиеся пропуски средним значением.

```python
ds['rating'].fillna(ds['rating'].mean(), inplace = True)
```

Опять проверим количество пропусков.

```python
ds.isnull().sum()
```

Видим, что количество пропусков теперь равно 0. Следовательно, можем анализировать датасет.

Для просмотра статистической сводки каждого столбца, чтобы узнать распределение данных в каждом столбце используется метод describe(). Этот метод показывает нам количество строк в столбце - count, среднее значение столбца - mean, столбец стандартное отклонение - std, минимальные (min) и максимальные (max) значения, а также границу каждого квартиля - 25%, 50% и 75%. Опишу свои наблюдения. Во-первых, обратимся к фичам заведений. Самая редкая из них: pets (среднее значение 0.000750), самая популярная: wi-fi (среднее значение 0.879506). Среднее значение рейтинга заведений Москвы и Санкт-Петербурга составляет 4.144113, минимальное значение рейтинга – 0, максимальное – 5. Отсюда следует, что заведения данных городов имеют относительно высокий рейтинг, что говорит о качественном обслуживании посетителей заведений общественного питания этих городов. Так как минимальный средний чек совпадает с медианой (квартиль 50%), то половина средних чеков всех заведений Москвы и Санкт-Петербурга составляет 500.
